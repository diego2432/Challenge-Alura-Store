### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda1 = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda1.head(5)
```

```python
tienda1.info()
```

#1. Análisis de facturación



##Ingreso total por cada tienda

###En este primer análisis, debes calcular el ingreso total de cada tienda. Sumando los valores de la columna Precio de cada conjunto de datos de la tienda para estimar los ingresos.

```python
ingreso_total_tienda1 = tienda1['Precio'].sum()
ingreso_total_tienda1
```

```python
ingreso_total_tienda2 = tienda2['Precio'].sum()
ingreso_total_tienda2
```

```python
ingreso_total_tienda3 = tienda3['Precio'].sum()
ingreso_total_tienda3
```

```python
ingreso_total_tienda4 = tienda4['Precio'].sum()
ingreso_total_tienda4
```

# 2. Ventas por categoría

###En este debes calcular la cantidad de productos vendidos por categoría en cada tienda. La idea es agrupar los datos por categoría y contar el número de ventas de cada tipo, mostrando las categorías más populares de cada tienda.

```python
categoria_tienda1 = tienda1['Categoría del Producto'].value_counts().rename('Productos_vendidos')
categoria_tienda1
```

```python
categoria_tienda2 = tienda2['Categoría del Producto'].value_counts().rename('Productos_vendidos')
categoria_tienda2
```

```python
categoria_tienda3 = tienda3['Categoría del Producto'].value_counts().rename('Productos_vendidos')
categoria_tienda3
```

```python
categoria_tienda4 = tienda4['Categoría del Producto'].value_counts().rename('Productos_vendidos')
categoria_tienda4
```

# 3. Calificación promedio de la tienda


###En este paso, debes calcular las calificaciones promedio de los clientes para cada tienda. El objetivo es conocer la satisfacción del cliente con los productos vendidos.

```python
valoracion_promedio_tienda1 = round(tienda1.Calificación.mean(),2)
valoracion_promedio_tienda1
```

```python
valoracion_promedio_tienda2 = round(tienda2.Calificación.mean(),2)
valoracion_promedio_tienda2
```

```python
valoracion_promedio_tienda3 = round(tienda3.Calificación.mean(),2)
valoracion_promedio_tienda3
```

```python
valoracion_promedio_tienda4 = round(tienda4.Calificación.mean(),2)
valoracion_promedio_tienda4
```

# 4. Productos más y menos vendidos

###En este paso, debes identificar qué productos fueron los más vendidos y los menos vendidos en cada tienda. Visualiza los resultados para que quede claro qué productos destacaron en ventas en cada tienda.

```python
productos_top_tienda1 = tienda1['Producto'].value_counts().rename('Unidades_vendidas').head(5)
productos_top_tienda1
```

```python
productos_top_tienda2 = tienda2['Producto'].value_counts().rename('Unidades_vendidas').head(5)
productos_top_tienda2
```

```python
productos_top_tienda3 = tienda3['Producto'].value_counts().rename('Unidades_vendidas').head(5)
productos_top_tienda3
```

```python
productos_top_tienda4 = tienda4['Producto'].value_counts().rename('Unidades_vendidas').head(5)
productos_top_tienda4
```

# 5. Envío promedio por tienda

###En este paso, debes calcular el costo de envío promedio para cada tienda. El objetivo es comprender cuánto se gasta, en promedio, en el envío de cada tienda.  Sabiendo que el cliente paga el envío.

```python
costo_envio_tienda1 = round(tienda1['Costo de envío'].mean(),2)
costo_envio_tienda1
```

```python
costo_envio_tienda2 = round(tienda2['Costo de envío'].mean(),2)
costo_envio_tienda2
```

```python
costo_envio_tienda3 = round(tienda3['Costo de envío'].mean(),2)
costo_envio_tienda3
```

```python
costo_envio_tienda4 = round(tienda4['Costo de envío'].mean(),2)
costo_envio_tienda4
```

#Visualización


```python
import matplotlib.pyplot as plt
```

```python
ventas_tiendas = [ingreso_total_tienda1,ingreso_total_tienda2,ingreso_total_tienda3,ingreso_total_tienda4]
tiendas = ['Tienda 1','Tienda 2','Tienda 3', 'Tienda 4']
ventas_billones = [venta / 1_000_000_000 for venta in ventas_tiendas]
etiquetas = [f'{tienda} - ${round(float(venta),2)}B' for tienda, venta in zip(tiendas, ventas_billones)]
plt.pie(
    ventas_tiendas,
    labels=etiquetas,
    autopct='%1.1f%%',
    startangle=90
    )
plt.axis('equal')
plt.title("Ventas de cada tienda porcentaje (%) y cantidad (Billones)")
plt.savefig('grafico_ventas.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
df = pd.concat([categoria_tienda1, categoria_tienda2, categoria_tienda3, categoria_tienda4],axis=1)

df.columns = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

plt.figure(figsize=(12,6))
df.plot(kind='bar')
plt.title('Cantidad de ventas por categoría en las 4 tiendas')
plt.xlabel('Categoría')
plt.ylabel('Cantidad vendida')
plt.xticks(rotation=90)
plt.legend(title="Tiendas")
plt.tight_layout()
plt.savefig('Ventas por categoria.png', dpi=300, bbox_inches='tight')
plt.show()

```

```python
valoraciones_tiendas = [valoracion_promedio_tienda1,valoracion_promedio_tienda2,
                        valoracion_promedio_tienda3,valoracion_promedio_tienda4]

plt.plot(tiendas, valoraciones_tiendas,marker="o",linewidth=2)
plt.title("Valoraciones por tienda")
plt.xlabel("Tiendas")
plt.ylabel("Calificación")
plt.ylim(3.95, 4.1)


plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.savefig('Valoraciones por tienda.png', dpi=300, bbox_inches='tight')
plt.show()
```
